```
# Simple Sentiment Analyzer
from textblob import TextBlob

print("🧠 Daily Mood Analyzer")
text = input("How are you feeling today? ")

analysis = TextBlob(text)
polarity = analysis.sentiment.polarity

if polarity > 0.1:
    mood = "Positive 😊"
elif polarity < -0.1:
    mood = "Negative 😔"
else:
    mood = "Neutral 😐"

print("\nSentiment score:", round(polarity, 3))
print("Mood:", mood)
```
