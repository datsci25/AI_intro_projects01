```
print("⚡ Personal Productivity Tracker\n")

study = float(input("Hours studied today: "))
coding = float(input("Hours coding today: "))
exercise = float(input("Hours exercising today: "))
scrolling = float(input("Hours on social media: "))

score = (study * 3) + (coding * 4) + (exercise * 2) - (scrolling * 2)

print("\n--- DAILY REPORT ---")
print("Study:", study)
print("Coding:", coding)
print("Exercise:", exercise)
print("Scrolling:", scrolling)

print("\nProductivity Score:", score)

if score >= 20:
    print("🔥 Extremely productive day!")
elif score >= 10:
    print("👍 Solid day.")
else:
    print("⚠️ Try to focus more tomorrow.")
```
